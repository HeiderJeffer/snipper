```python
#!/usr/bin/env python3
"""
Interactive Projectile Simulator (Educational)

Features:
- 2D projectile with quadratic drag, steady wind, gravity.
- CLI prompts for inputs (with defaults).
- Options: single sim, angle sweep (max range), target-hitting sweep.
- Optional terrain slope and target height.
- Optional CSV export and plotting (matplotlib).

Dependencies:
- numpy (recommended)
- matplotlib (optional; for plotting)
- If numpy is missing the script will still run but slower (uses math only).

Author: ChatGPT (educational example)
"""

from dataclasses import dataclass
import math
import csv
import sys

# Try to import numpy; fall back gracefully.
try:
    import numpy as np
    HAS_NUMPY = True
except Exception:
    HAS_NUMPY = False

# Constants
g = 9.80665  # m/s^2

@dataclass
class ProjectileParams:
    speed: float
    angle_deg: float
    height: float = 0.0
    mass: float = 0.145
    area: float = 0.0042
    cd: float = 0.47
    air_density: float = 1.225
    wind_speed: float = 0.0
    dt: float = 0.002
    max_time: float = 60.0
    ground_slope_deg: float = 0.0  # slope of terrain (deg); 0 = flat
    ground_offset: float = 0.0     # y-intercept of ground (m)

@dataclass
class Result:
    time: float
    range: float
    max_height: float
    impact_x: float
    impact_y: float
    trace: list  # list of (t,x,y,vx,vy)

def terrain_height_at(x, params: ProjectileParams):
    # Ground is a line: y = m*x + b, where m = tan(slope)
    m = math.tan(math.radians(params.ground_slope_deg))
    return m * x + params.ground_offset

def drag_accel(vx, vy, params: ProjectileParams):
    # Relative velocity to air (wind along +x)
    vrel_x = vx - params.wind_speed
    vrel_y = vy
    vrel = math.hypot(vrel_x, vrel_y)
    if vrel == 0:
        return 0.0, 0.0
    D = 0.5 * params.air_density * params.cd * params.area * vrel * vrel
    ax = -D * (vrel_x / vrel) / params.mass
    ay = -D * (vrel_y / vrel) / params.mass
    return ax, ay

def rk4_step(x, y, vx, vy, dt, params: ProjectileParams):
    def deriv(vx, vy):
        ax_d, ay_d = drag_accel(vx, vy, params)
        return vx, vy, ax_d, ay_d - g

    k1_vx, k1_vy, k1_ax, k1_ay = deriv(vx, vy)
    k2_vx, k2_vy, k2_ax, k2_ay = deriv(vx + 0.5*dt*k1_ax, vy + 0.5*dt*k1_ay)
    k3_vx, k3_vy, k3_ax, k3_ay = deriv(vx + 0.5*dt*k2_ax, vy + 0.5*dt*k2_ay)
    k4_vx, k4_vy, k4_ax, k4_ay = deriv(vx + dt*k3_ax, vy + dt*k3_ay)

    x_next  = x  + dt*(k1_vx + 2*k2_vx + 2*k3_vx + k4_vx)/6.0
    y_next  = y  + dt*(k1_vy + 2*k2_vy + 2*k3_vy + k4_vy)/6.0
    vx_next = vx + dt*(k1_ax + 2*k2_ax + 2*k3_ax + k4_ax)/6.0
    vy_next = vy + dt*(k1_ay + 2*k2_ay + 2*k3_ay + k4_ay)/6.0
    return x_next, y_next, vx_next, vy_next

def simulate(params: ProjectileParams, stop_at_ground=True, include_trace=True) -> Result:
    theta = math.radians(params.angle_deg)
    vx = params.speed * math.cos(theta)
    vy = params.speed * math.sin(theta)
    t = 0.0
    x = 0.0
    y = params.height
    trace = []
    max_h = y

    while t < params.max_time:
        if include_trace:
            trace.append((t, x, y, vx, vy))
        max_h = max(max_h, y)
        x_next, y_next, vx_next, vy_next = rk4_step(x, y, vx, vy, params.dt, params)
        t_next = t + params.dt

        if stop_at_ground:
            # Check if crossed the terrain between (x,y) and (x_next,y_next)
            y_ground = terrain_height_at(x, params)
            y_ground_next = terrain_height_at(x_next, params)
            if y_next <= y_ground_next:
                # interpolate to approximate impact point where y == terrain_height_at(x)
                # We do linear interpolation in time for y - ground(x)
                f1 = y - y_ground
                f2 = y_next - y_ground_next
                if f1 == f2:
                    alpha = 0.0
                else:
                    alpha = f1 / (f1 - f2)
                t_hit = t + alpha * (t_next - t)
                x_hit = x + alpha * (x_next - x)
                y_hit = y + alpha * (y_next - y)
                # adjust final trace
                if include_trace:
                    trace.append((t_hit, x_hit, y_hit, vx + alpha*(vx_next - vx), vy + alpha*(vy_next - vy)))
                return Result(time=t_hit, range=x_hit, max_height=max_h, impact_x=x_hit, impact_y=y_hit, trace=trace)
        # step
        x, y, vx, vy, t = x_next, y_next, vx_next, vy_next, t_next

        # safety: if projectile goes very far/falls indefinitely, break
        if x > 1e6 or abs(y) > 1e6:
            break

    # if we exit loop without hitting ground
    return Result(time=t, range=x, max_height=max_h, impact_x=x, impact_y=y, trace=trace)

def sweep_max_range(params: ProjectileParams, angle_min=0.1, angle_max=89.9, steps=181):
    best = None
    best_angle = None
    for a in np.linspace(angle_min, angle_max, steps) if HAS_NUMPY else [angle_min + i*(angle_max-angle_min)/(steps-1) for i in range(steps)]:
        p = ProjectileParams(**{**params.__dict__, "angle_deg": float(a)})
        res = simulate(p)
        if best is None or res.range > best.range:
            best = res
            best_angle = a
    return best_angle, best

def find_angle_to_hit_target(params: ProjectileParams, target_x, target_y, angle_low=1.0, angle_high=88.0, tol=0.5, max_iter=40):
    # Try to find an angle that lands within 'tol' meters of (target_x, target_y).
    # We'll use a sampled search + simple local refinement since the function is not strictly monotonic in drag.
    if not HAS_NUMPY:
        angles = [angle_low + i*(angle_high-angle_low)/99.0 for i in range(100)]
    else:
        angles = np.linspace(angle_low, angle_high, 201)

    best_angle = None
    best_dist = float("inf")
    best_result = None

    for a in angles:
        p = ProjectileParams(**{**params.__dict__, "angle_deg": float(a)})
        res = simulate(p)
        dx = res.impact_x - target_x
        dy = res.impact_y - target_y
        dist = math.hypot(dx, dy)
        if dist < best_dist:
            best_dist = dist
            best_angle = a
            best_result = res
            if best_dist <= tol:
                break

    return best_angle, best_dist, best_result

def prompt_float(prompt_text, default):
    try:
        s = input(f"{prompt_text} [{default}]: ").strip()
        if s == "":
            return float(default)
        return float(s)
    except Exception:
        print("Invalid input; using default.")
        return float(default)

def prompt_yesno(prompt_text, default=True):
    d = "Y/n" if default else "y/N"
    s = input(f"{prompt_text} ({d}): ").strip().lower()
    if s == "":
        return default
    return s.startswith("y")

def save_trace_csv(trace, filename):
    try:
        with open(filename, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["t","x","y","vx","vy"])
            for row in trace:
                writer.writerow(row)
        print(f"Wrote CSV to: {filename}")
    except Exception as e:
        print("Failed to write CSV:", e)

def try_plot(trace, terrain_params=None, target=None):
    try:
        import matplotlib.pyplot as plt
    except Exception:
        print("matplotlib not available; skipping plot.")
        return
    t = [r[0] for r in trace]
    x = [r[1] for r in trace]
    y = [r[2] for r in trace]
    plt.figure(figsize=(8,5))
    plt.plot(x, y, label="trajectory")
    if terrain_params is not None:
        xmin = min(0, min(x))
        xmax = max(x)
        xs = [xmin + i*(xmax-xmin)/200 for i in range(201)]
        ys = [terrain_height_at(xx, terrain_params) for xx in xs]
        plt.plot(xs, ys, label="terrain")
    if target is not None:
        tx, ty = target
        plt.scatter([tx],[ty], marker="x", s=80, label="target")
    plt.xlabel("x (m)")
    plt.ylabel("y (m)")
    plt.legend()
    plt.grid(True)
    plt.title("Projectile Trajectory")
    plt.show()

def run_interactive():
    print("=== Projectile Simulator (Educational) ===")
    print("Enter parameters (press Enter to accept default shown in brackets).")
    # Basic params
    speed = prompt_float("Launch speed (m/s)", 30.0)
    angle = prompt_float("Launch angle (degrees)", 35.0)
    height = prompt_float("Launch height (m)", 1.8)
    mass = prompt_float("Mass (kg)", 0.145)
    area = prompt_float("Cross-sectional area (m^2)", 0.0042)
    cd = prompt_float("Drag coefficient (Cd)", 0.47)
    rho = prompt_float("Air density (kg/m^3)", 1.225)
    wind = prompt_float("Wind speed (m/s, + = tailwind)", 0.0)
    dt = prompt_float("Time step dt (s) - smaller -> more accurate, default 0.002", 0.002)
    max_time = prompt_float("Max sim time (s)", 60.0)
    slope = prompt_float("Terrain slope (degrees, positive = upward to +x)", 0.0)
    ground_offset = prompt_float("Terrain vertical offset (m) (y-intercept at x=0)", 0.0)

    params = ProjectileParams(
        speed=speed, angle_deg=angle, height=height, mass=mass, area=area, cd=cd,
        air_density=rho, wind_speed=wind, dt=dt, max_time=max_time,
        ground_slope_deg=slope, ground_offset=ground_offset
    )

    print("\nChoose action:")
    print("1) Single simulation")
    print("2) Sweep angles to find maximum range")
    print("3) Find angle to hit a target (search)")
    choice = input("Select 1/2/3 [1]: ").strip() or "1"

    if choice == "1":
        res = simulate(params)
        print(f"\n--- Results ---")
        print(f"Flight time: {res.time:.4f} s")
        print(f"Impact (x,y): ({res.impact_x:.3f} m, {res.impact_y:.3f} m)")
        print(f"Horizontal range (x): {res.range:.3f} m")
        print(f"Max height: {res.max_height:.3f} m")
        if prompt_yesno("Save trajectory CSV?", False):
            fn = input("Filename [trajectory.csv]: ").strip() or "trajectory.csv"
            save_trace_csv(res.trace, fn)
        if prompt_yesno("Show plot (requires matplotlib)?", True):
            try_plot(res.trace, terrain_params=params)
    elif choice == "2":
        steps = int(prompt_float("Number of angle samples (higher = slower, e.g. 181)", 181))
        ang_min = prompt_float("Angle min (deg)", 0.1)
        ang_max = prompt_float("Angle max (deg)", 89.9)
        if not HAS_NUMPY:
            print("Note: numpy not installed; sweep will still run but be slower.")
        best_angle, best_res = sweep_max_range(params, angle_min=ang_min, angle_max=ang_max, steps=steps)
        print(f"\nBest angle (sampled): {best_angle:.4f} deg")
        print(f"Range at best angle: {best_res.range:.3f} m")
        print(f"Flight time: {best_res.time:.4f} s, Max height: {best_res.max_height:.3f} m")
        if prompt_yesno("Save best trajectory CSV?", False):
            fn = input("Filename [best_trajectory.csv]: ").strip() or "best_trajectory.csv"
            save_trace_csv(best_res.trace, fn)
        if prompt_yesno("Show plot of best trajectory?", True):
            try_plot(best_res.trace, terrain_params=params)
    elif choice == "3":
        tx = prompt_float("Target x (m)", 50.0)
        ty = prompt_float("Target y (m)", 0.0)
        tol = prompt_float("Acceptable miss distance (m)", 1.0)
        print("Searching for an angle that lands near the target... (this is a sampled search)")
        best_angle, best_dist, best_res = find_angle_to_hit_target(params, tx, ty, angle_low=1.0, angle_high=88.0, tol=tol)
        if best_angle is None:
            print("No candidate angle found.")
        else:
            print(f"Best angle found: {best_angle:.4f} deg (miss distance {best_dist:.3f} m)")
            print(f"Impact at ({best_res.impact_x:.3f}, {best_res.impact_y:.3f}) after {best_res.time:.3f} s")
            if prompt_yesno("Save this trajectory CSV?", False):
                fn = input("Filename [target_trajectory.csv]: ").strip() or "target_trajectory.csv"
                save_trace_csv(best_res.trace, fn)
            if prompt_yesno("Show plot (trajectory + terrain + target)?", True):
                try_plot(best_res.trace, terrain_params=params, target=(tx,ty))
    else:
        print("Unknown choice. Exiting.")

if __name__ == "__main__":
    run_interactive()
```
